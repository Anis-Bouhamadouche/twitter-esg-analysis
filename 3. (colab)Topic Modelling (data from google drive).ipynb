```python
drive.mount('/content/drive/')
data_path = '/content/drive/MyDrive/Mini-Project/'
```

```python
def load_tweets(path = data_path):
    tweets = []
    failed_loading = []
    count = 0
    while count <= 5000:
        try:
            tweets_df = pd.read_csv(f'{path}tweet_{count}.csv')
            tweets.append(tweets_df)
        except:
            print(f"tweets set {count} failed to load")
            failed_loading.append(count)
        count += 1
    
    tweets_df = pd.concat(tweets)
    
    return tweets_df, failed_loading
```

```python
tweets_df, failed_loading = load_tweets()
```

```python
tweets_df.drop(columns='Unnamed: 0', inplace = True)
```

```python
tweets_df.dropna(subset =['text'], inplace = True)
```

# Topic modelling

```python
from top2vec import Top2Vec
```

```python
model = Top2Vec(list(tweets_df.text[:500000]), embedding_model='all-MiniLM-L6-v2')
```

```python
model.get_num_topics()
```
